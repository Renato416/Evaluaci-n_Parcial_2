### FASE 1: Limpieza de Datos (Data Cleaning)

1.1 Eliminar customerID

Este código elimina la columna customerID del dataset. Esta variable corresponde a un identificador único de cada cliente y no aporta información útil para el modelo de predicción. Mantenerla podría generar ruido en el entrenamiento de la inteligencia artificial, ya que el modelo podría intentar encontrar patrones en valores aleatorios.

```python
print("\n--- ELIMINAR COLUMNA customerID ---")
df = df.drop("customerID", axis=1)
display(df.head())
```

La limpieza de datos fue realizada correctamente. La columna customerID ya no aparece en el dataset, lo que confirma que fue eliminada exitosamente. Este paso es importante porque los identificadores únicos no aportan valor predictivo al modelo de inteligencia artificial y podrían introducir ruido durante el entrenamiento.

Al revisar el head() del dataframe, se observa que ahora solamente permanecen las variables relevantes para el análisis del churn, incluyendo datos demográficos, servicios contratados, facturación y la variable objetivo Churn.

1.2 Corregir tipo de dato de TotalCharges
Este código convierte la columna TotalCharges desde tipo texto (object) a tipo numérico (float). Esto es necesario porque la columna contiene valores que parecen números, pero fueron interpretados como texto al cargar el CSV. Además, se utiliza errors='coerce' para transformar valores vacíos o inválidos en NaN, permitiendo detectar datos faltantes ocultos.

```python
print("\n--- CORREGIR TIPO DE DATO DE TotalCharges ---")
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
print("\nTipo de dato actual:")
print(df["TotalCharges"].dtype)
print("\nPrimeros valores de TotalCharges:")
display(df["TotalCharges"].head())
```

La conversión de la columna TotalCharges fue realizada correctamente. El tipo de dato cambió exitosamente a float64, lo que confirma que ahora los valores pueden ser utilizados en operaciones matemáticas, análisis estadísticos y modelos de Machine Learning.

Además, los primeros registros muestran valores numéricos válidos, indicando que la columna ya se encuentra en un formato adecuado para continuar con el procesamiento de datos. Hasta este punto no se observan valores NaN en los primeros registros visualizados, aunque aún es necesario verificar si existen datos faltantes en otras filas del dataset.

1.3 Manejo de valores nulos

Este código revisa la cantidad de valores nulos presentes en cada columna del dataset. Luego, reemplaza los posibles valores faltantes (NaN) de la columna TotalCharges por 0, evitando problemas durante el entrenamiento de modelos de inteligencia artificial.

```python
print("\n--- REVISIÓN DE VALORES NULOS ---")
display(df.isnull().sum())
```

```python
print("\n--- REEMPLAZAR VALORES NULOS EN TotalCharges ---") # habian 11 nulos en TotalCharges
df["TotalCharges"] = df["TotalCharges"].fillna(0)
print("\n--- REVISIÓN FINAL DE NULOS ---")
display(df.isnull().sum())
```

La revisión de valores nulos permitió identificar 11 registros vacíos en la columna TotalCharges. Estos valores faltantes fueron reemplazados exitosamente por 0 utilizando fillna(0).

Después de la limpieza, el dataset quedó completamente libre de valores nulos en todas sus columnas, lo que garantiza una mejor calidad de los datos para las siguientes etapas del proyecto. Este paso es fundamental, ya que la mayoría de los algoritmos de Machine Learning no pueden trabajar correctamente con datos vacíos.

1.4 asegurar que no haya filas clonadas

```python
df.drop_duplicates()
```

Al eliminar la columna customerID (que hacía que cada fila fuera única), quedaron expuestos 22 clientes que tienen exactamente el mismo perfil: mismo género, misma edad, mismos servicios contratados, mismos meses de antigüedad (tenure) y exactamente los mismos cargos mensuales y totales.

1.4.1 Búsqueda y eliminación de duplicados

```python
print("\n--- BÚSQUEDA DE DUPLICADOS ---")
duplicados = df.duplicated().sum()
print(f"Filas duplicadas encontradas: {duplicados}")

print("\n--- ELIMINAR DUPLICADOS ---")
df = df.drop_duplicates()
print(f"Tamaño del dataset tras eliminar duplicados: {df.shape}")
```

La revisión de duplicados permitió identificar 22 filas repetidas dentro del dataset. Esto ocurrió porque, al eliminar la columna customerID, algunos registros quedaron con exactamente las mismas características en todas sus variables restantes.

Posteriormente, los registros duplicados fueron eliminados exitosamente utilizando drop_duplicates(), reduciendo el tamaño del dataset de 7043 filas a 7021 filas.

### FASE 2: Análisis Exploratorio

2.1 Análisis Bivariado: Contract vs Churn

Este código genera un gráfico de barras para comparar el tipo de contrato (Contract) con la variable objetivo Churn. El gráfico permitirá observar y evaluar qué tipo de contrato presenta una mayor cantidad de clientes que abandonan el servicio.

En el gráfico:

- No aparece representado en color azul.
- Yes aparece representado en color naranja.

Esto ayudará a identificar visualmente si los clientes con contratos mensuales (Month-to-month) tienen una mayor tendencia al abandono en comparación con los contratos de 1 o 2 años.

```python
print("\n--- GRÁFICO CONTRACT VS CHURN ---")
plt.figure(figsize=(10,6))
sns.countplot(
    data=df,
    x="Contract",
    hue="Churn"
)
plt.title("Contract vs Churn")
plt.xlabel("Tipo de Contrato")
plt.ylabel("Cantidad de Clientes")
plt.show()
```

El gráfico muestra que los clientes con contrato Month-to-month presentan la mayor cantidad de abandono del servicio (Churn = Yes, color naranja). Además, aunque este tipo de contrato también concentra muchos clientes que permanecen en la compañía (No), la diferencia entre ambos grupos es considerablemente menor en comparación con los contratos de largo plazo.

Por otro lado, los contratos de One year y Two year presentan una cantidad mucho menor de clientes que abandonan el servicio, lo que indica que los contratos de mayor duración están asociados a una mayor retención de clientes.

Este análisis sugiere que el tipo de contrato es una variable importante para predecir el churn, especialmente en clientes con contratos mensuales.

2.2 Análisis Bivariado: tenure vs Churn

Este código genera un gráfico para analizar la relación entre la antigüedad del cliente (tenure) y el abandono del servicio (Churn). El objetivo es observar cómo se comporta la permanencia de los clientes y evaluar si los clientes más nuevos tienen una mayor tendencia a abandonar la compañía

```python
print("\n--- GRÁFICO TENURE VS CHURN ---")

plt.figure(figsize=(12,6))

sns.histplot(
    data=df,
    x="tenure",
    hue="Churn",
    multiple="stack",
    bins=30
)

plt.title("Tenure vs Churn")
plt.xlabel("Meses de permanencia (tenure)")
plt.ylabel("Cantidad de Clientes")

plt.show()
```
